# Cifra de Vigenère

```python
# Conceito: escolher uma chave e repetí-la até que ela seja do mesmo tamanho que a palavra a ser cifrada;
# então, usar uma matriz de caracteres do alfabeto para então cifrar o texto, aplicando uma cifra de Cesar a cada letra.
# Afim de realizar isso de forma matemática é possivel utilizando as fórmulas a seguir,
# onde P é a letra da palavra, K é a letra da chave, N é o número de caracteres do alfabeto e C é a letra cifrada.
# C = P + K , MOD N
# P = C - K + N, MOD N
```

```python
# definimos uma função para  calcular o módulo de um número
```

```python
def myMod(value, MOD): 
    return value % MOD
```

```python
# definimos N, o número de cacteres do alfabeto com o máximo permitido pela linguagem
```

```python
import sys
n = sys.maxunicode
```

```python
message = input("Enter a message: ")
```

```python
key = input("Enter a key: ")
```

```python
# replicamos a key afim de deixá-la com o mesmo tamanho da mensagem
```

```python
if len(message) > len(key):
    while len(message) > len(key):
        key= key +key
```

```python
key = key[0:len(message)]
```

```python
# uma visualização dos dados atuais
```

```python
print(message)
print(key)
```

```python
# afim de criptografar a mensagem, usamos a formula C = (P + K) , MOD N
```

```python
cifra =""
i = 0
for letter in message:    
    p = ord(letter)
    k = ord(key[i])
    value = p+k    
    cifra = cifra + (chr(myMod(value,n)))
```

```python
# mensagem criptografada
```

```python
print(cifra)
```

```python
# afim de obter a mensagem original (decriptografada), usamos a outra fórmula 
```

```python
decp =""
i = 0
for letter in cifra:    
    c = ord(letter)
    k = ord(key[i])
    value = c-k +n
    decp = decp + (chr(myMod(value,n)))
```

```python
# mensagem resultante do processo:
```

```python
print(decp)
```

```python
# uma pequena verificação entre a mensagem obitida e o resultado final
```

```python
print (decp == message)
```
